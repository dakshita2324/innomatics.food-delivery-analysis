```python
import pandas as pd
import numpy as np
```

```python
df_csv = pd.read_csv("orders.csv")
df_csv.head()
```

```python
df_json = pd.read_json("users.json", orient = "records")
```

```python
import sqlite3
conn = sqlite3.connect("restaurants.db")
with open("restaurants.sql") as f:
    sql_script = f.read()
conn.executescript(sql_script)
```

```python
restaurants = pd.read_sql_query("SELECT * FROM restaurants",conn)
restaurants.head()
```

```python
orders_users_df = pd.merge(
    df_csv,
    df_json,
    on="user_id",
    how = "left"
)
orders_users_df.head()
```

```python
final_df = pd.merge(
    orders_users_df,
    restaurants,
    on = "restaurant_id",
    how = "left"
)
final_df.head()
```

```python
final_df.shape
```

```python
final_df.columns
```

```python
final_df = final_df.drop(columns=["restaurants_name_y"],errors = "ignore")
final_df = final_df.rename(columns={"restaurants_name_x":"rstaurant_name"})
```

```python
final_df.to_csv("final_food_delivery_dataset.csv",index = False)
```

```python
options = [
    "Grand Cafe Punjabi",
    "Grand Restaurant South India",
    "Ruchi Mess Multicuisine",
    "Ruchi Foods Chinese"
]

q8 = (
    final_df[final_df["restaurant_name_x"].isin(options)]
    .groupby("restaurant_name_x")
    .agg(
        avg_order=("total_amount", "mean"),
        order_count=("order_id", "count")
    )
    .query("order_count < 20")
    .sort_values("avg_order", ascending=False)
    .head(1)
)

q8
```

```python
# Find top revenue city among Gold members
top_gold_city = (
    final_df[final_df["membership"] == "Gold"]
    .groupby("city")["total_amount"]
    .sum()
    .idxmax()
)

# Count orders in that city by Gold members
orders_top_city = final_df[
    (final_df["membership"] == "Gold") &
    (final_df["city"] == top_gold_city)
].shape[0]

top_gold_city, orders_top_city
```
