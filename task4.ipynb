```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        print("Recharge amount must be at least ₹50.")
        return False
    elif amount not in valid_plans:
        print("Invalid plan. Choose from 199, 299, 399, 599.")
        return False
    else:
        print("Recharge successful!")
        return True

while True:
    amount = int(input("Enter recharge amount: "))
    if validate_recharge(amount):
        break
```

```python
def check_stock(products):
    for item, quantity in products.items():
        if quantity < 15:
            print(item, ": Reorder Alert")
        else:
            print(item, ": Stock OK")


# Example dictionary
inventory = {
    "Rice": 10,
    "Sugar": 25,
    "Oil": 12,
    "Salt": 20
}

check_stock(inventory)
```

```python
def process_result(marks):
    total = 0

    # Calculate total using loop
    for m in marks:
        total += m

    average = total / len(marks)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"


# Example
student_marks = [60, 45, 70, 55]
result = process_result(student_marks)

print("Result:", result)
```

```python
def calculate_fare(distance, peak):
    base_fare = 50
    per_km = 12

    fare = base_fare + (distance * per_km)

    if peak == "yes":
        fare += fare * 0.25   # 25% extra

    return fare


# Retry using while loop
while True:
    distance = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ").lower()

    total_fare = calculate_fare(distance, peak)
    print("Total Fare: ₹", total_fare)

    retry = input("Do you want to calculate again? (yes/no): ").lower()
    if retry != "yes":
        break
```

```python
def check_eligibility(attendance):
    present_days = 0

    # Count present days
    for day in attendance:
        if day == "P":
            present_days += 1

    percentage = (present_days / len(attendance)) * 100

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

attendance_list = ["P", "A", "P", "P", "A", "P", "P", "P"]
result = check_eligibility(attendance_list)

print("Attendance Percentage:", (attendance_list.count("P") / len(attendance_list)) * 100)
print("Result:", result)
```

```python
def check_password(password):
    has_digit = False
    has_special = False

    # Check each character
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in "@#$":
            has_special = True

    if len(password) >= 8 and has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"


# Example
pwd = input("Enter password: ")
result = check_password(pwd)
print(result)
```
