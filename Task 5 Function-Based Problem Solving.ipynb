### 1.Social Media – Post Engagement Analyzer

```python
def analyze_engagement(likes_list):
    total_likes = 0
    for like in likes_list:
        total_likes += like
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    print("Total Likes:", total_likes)
    print("Post Status:", status)
likes = [200, 150, 300, 250, 150]
analyze_engagement(likes)
```

### 3.Agriculture – Rainfall Adequacy Checker

```python
def check_rainfall(rainfall_list, required_level):
    total = 0
    for rain in rainfall_list:
        total += rain
    average = total // len(rainfall_list)
    print("Average Rainfall:", average)
    if average >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")
rainfall_data = [70, 80, 65, 75, 70]
required = 70
check_rainfall(rainfall_data, required)
```

### 4.Social Media – Duplicate Account Detection

```python
def detect_duplicate_accounts(usernames):
    unique_users = set(usernames)
    if len(unique_users) != len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")
users = ["ram123", "sita45", "john77", "ram123", "alex9"]
detect_duplicate_accounts(users)
```

### 7.System Monitoring – Application Health Checker

```python
def check_application_health(errors):
    print("Error Count:", errors)
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    print("System Status:", status)
error_count = 7
check_application_health(error_count)
```

### 9.E-Learning – Student Attendance Eligibility System

```python
def check_exam_eligibility(attendance_list):
    total_days = len(attendance_list)
    present_days = 0
    for day in attendance_list:
        if day == "P":
            present_days += 1
    percentage = (present_days / total_days) * 100
    print("Attendance Percentage:", percentage)
    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")
attendance = ["P", "P", "A", "P", "P"]
check_exam_eligibility(attendance)
```

### 10.Smart Electricity Bill Analyzer

```python
def calculate_electricity_bill(units):
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"
    return bill, status
units_consumed = 250
total_bill, usage_status = calculate_electricity_bill(units_consumed)
print("Units Consumed:", units_consumed)
print("Total Bill:", total_bill)
print("Usage Level:", usage_status)
```
